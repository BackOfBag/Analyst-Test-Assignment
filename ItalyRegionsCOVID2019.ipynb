Устанавливаем стандартные библиотеки и загружаем данные:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline

warnings.filterwarnings(action="ignore")

df = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv")
df
```

Описание и формат данных [приведены](https://github.com/pcm-dpc/COVID-19/blob/master/dati-andamento-covid19-italia.md) по ссылке.

```python
info = pd.DataFrame(df.dtypes)
info = info.rename(columns={0:'Тип данных'})
info['Количество'] = df.count()
info['Уникальных Значений'] = df.nunique()
info['Пропущенных Значений'] = df.isnull().sum()
info
```

Нас интересуют данные по выявленным случаям COVID-19 (**totale_positivi**) и умершим от COVID-19 (**deceduti**) для области Ломбардия в Италии (**Lombardia**):

```python
Lombardia = df[['data', 'denominazione_regione', 'totale_positivi', 'deceduti']].loc[df.denominazione_regione == 'Lombardia']
Lombardia
```

Численность населения Италии на 01.02.2020г. согласно [статистическим данным](https://esploradati.istat.it/databrowser/#/en/dw/dashboards):

```python
Italy_Population_Feb2020 = 59_614_281
```

Уточним выявленные положительные случаи COVID-19 и количество умерших от COVID-19 в пересчете на 1 млн человек:

```python
Lombardia['totale_positivi_per_mln'] = Lombardia.totale_positivi * 1_000_000 / Italy_Population_Feb2020
Lombardia['deceduti_per_mln'] = Lombardia.deceduti * 1_000_000 / Italy_Population_Feb2020
Lombardia = Lombardia.round(2)
Lombardia
```

```python
plt.figure(figsize=(12, 5))
plt.title("Положительные случаи COVID-19 на 1 млн человек в провинции Ломбардия, Италия")
sns.lineplot(data=Lombardia, x='data', y='totale_positivi_per_mln')
```

```python
plt.figure(figsize = (12, 5))
plt.title("Количество смертей от COVID-19 на 1 млн человек в провинции Ломбардия, Италия")
sns.lineplot(data=Lombardia, x='data', y='deceduti_per_mln')
```
